# Visualización de los datos

```python
plt.figure(figsize=(15,15))
plt.title("Distribuciones de las caracteristicas")
plt.subplots_adjust(wspace=0.2, hspace=0.3) 
plt.axis('off')
for i, col in enumerate(df_cardiovascualar.drop(columns=["patientid"]).columns):
    plt.subplot(5,3,i+1)
    sns.histplot(x=col, data=df_cardiovascualar.drop(columns=["patientid"]),hue='target', kde=True)
    plt.ylabel("Conteos")
```

```python
df_cardiovascualar_corr = df_cardiovascualar.corr()
df_cardiovascualar_corr
```

```python
plt.figure(figsize = (20, 20)) 
sns.heatmap(data=round(df_cardiovascualar.corr(), 2), annot= True,cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize = (20, 20)) 
sns.pairplot(df_cardiovascualar)
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.title("Diagramas de cajas y bigotes")
plt.subplots_adjust(wspace=0.2, hspace=0.3) 
plt.axis('off')
for i, col in enumerate(df_cardiovascualar.drop(columns=["patientid"]).columns):
    plt.subplot(5,3,i+1)
    sns.boxplot(x=col, data=df_cardiovascualar.drop(columns=["patientid"]))
```

```python
col1='serumcholestrol'
col2='oldpeak'
fig = px.scatter(
    df_cardiovascualar, x=col1, y=col2,
    color='target',
    title=f'Heart disea: {col1} vs {col2}',
    labels={'target': 'Diagnosis'}
)
fig.show()
```
